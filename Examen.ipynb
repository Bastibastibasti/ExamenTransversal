```python
run = True
from tabulate import tabulate
from datetime import datetime
fecha_actual = datetime.now()

edificio = [
    ["1a", "Disponible"], ["1b", "Disponible"], ["1c", "Disponible"], ["1d", "Disponible"],
    ["2a", "Disponible"], ["2b", "Disponible"], ["2c", "Disponible"], ["2d", "Disponible"],
    ["3a", "Disponible"], ["3b", "Disponible"], ["3c", "Disponible"], ["3d", "Disponible"],
    ["4a", "Disponible"], ["4b", "Disponible"], ["4c", "Disponible"], ["4d", "Disponible"],
    ["5a", "Disponible"], ["5b", "Disponible"], ["5c", "Disponible"], ["5d", "Disponible"],
    ["6a", "Disponible"], ["6b", "Disponible"], ["6c", "Disponible"], ["6d", "Disponible"],
    ["7a", "Disponible"], ["7b", "Disponible"], ["7c", "Disponible"], ["7d", "Disponible"],
    ["8a", "Disponible"], ["8b", "Disponible"], ["8c", "Disponible"], ["8d", "Disponible"],
    ["9a", "Disponible"], ["9b", "Disponible"], ["9c", "Disponible"], ["9d", "Disponible"],
    ["10a", "Disponible"], ["10b", "Disponible"], ["10c", "Disponible"], ["10d", "Disponible"]
]

def menu():
  print("INMOBILIARIA MI CASA\n")
  print("Para ver los departamentos disponibles presione 1")
  print("Para comprar un departamento presione 2")
  print("Para ver el listado de compradores presione 3")
  print("Para ver las ganancias totales presione 4")
  print("Para salir presione 5")
  opcion=input("Seleccione una opcion : ")
  if opcion == "1" :
    mostrar_disponibles()
  elif opcion== "2" :
      comprar()
  elif opcion=="3" :
    buscar_compradores()
  elif opcion =="4" :
    mostrar_ganancias()
  elif opcion=="5":
    print("Muchas gracias, hasta luego\nDesarrollador: Bastian Herrera \n", fecha_actual)
    global run
    run = False
  else :
    print("Opcion incorrecta, vuelva a ingresar su opcion")
    run = False
def mostrar_disponibles():
    print("\t\t\tLISTADO DE DEPARTAMENTOS")
    from tabulate import tabulate
    headers = ["Piso"] + [letra for letra in "ABCD"]
    tabla = []
    for piso in range(9, -1, -1):
        fila = [str(piso + 1)] + [departamento[1] for departamento in edificio[piso * 4: (piso + 1) * 4]]
        tabla.append(fila)
    print(tabulate(tabla, headers=headers, tablefmt="fancy_grid"))

def comprar ():
  print("COMPRAR DEPARTAMENTO")
  nombre=input("Ingrese su nombre: ")
  apellido=input("Ingrese su apellido: ")
  rut=input("Ingrese su rut: ")
  piso=input("Ingrese el piso donde desea comprar su departamento: ")
  tipo_departamento=input("Ingrese el tipo de departamento que desea comprar; A,B,C o D: ")
  eleccion = str(piso) + tipo_departamento.lower()
  for departamento in edificio :
    if eleccion == departamento[0] :
      if departamento[1]== "Disponible":
        departamento[1] = "VENDIDO"
        datos = (nombre, apellido, rut)
        departamento.append(datos)
        print("Â¡Departamento comprado con exito!")
      return
      print("Departamento no disponible, por favor seleccione otro")

def buscar_compradores():
    print("LISTADO DE COMPRADORES")
    compradores = []
    for departamento in edificio:
        if len(departamento) > 2 and departamento[1] == "VENDIDO":
            compradores.append([departamento[2][0], departamento[2][1], departamento[2][2], departamento[0], departamento[1]])
    if compradores:
        headers = ["Nombre", "Apellido", "RUT", "Departamento", "Estado"]
        print(tabulate(compradores, headers=headers, tablefmt="fancy_grid"))
    else:
        print("No se encontraron compradores.")

from tabulate import tabulate

def mostrar_ganancias():
    compraa = 0
    comprab = 0
    comprac = 0
    comprad = 0
    print("\nEL TOTAL RECAUDADO ES: ")
    for departamento in edificio:
        if departamento[1] == "VENDIDO":
            tipo_departamento = departamento[0][-1]
            if tipo_departamento == "a":
                compraa += 1
            elif tipo_departamento == "b":
                comprab += 1
            elif tipo_departamento == "c":
                comprac += 1
            elif tipo_departamento == "d":
                comprad += 1

    datos = [
        ["Tipo de Departamento", "Cantidad Vendida", "Valor Recaudado"],
        ["A", compraa, compraa * 3800],
        ["B", comprab, comprab * 3000],
        ["C", comprac, comprac * 2800],
        ["D", comprad, comprad * 3500]
    ]

    headers = datos.pop(0)
    print(tabulate(datos, headers=headers, tablefmt="fancy_grid"))





while run :
  menu()
```
